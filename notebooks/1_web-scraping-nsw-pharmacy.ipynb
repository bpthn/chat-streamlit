```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.health.nsw.gov.au/pharmaceutical/Pages/pharmacy-trial-locations.aspx"
response = requests.get(url)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table')
    headers = [header.text.strip() for header in table.find_all('th')]
    
    rows = []
    for row in table.find_all('tr')[1:]:
        rows.append([data.text.strip() for data in row.find_all('td')])
    
    df = pd.DataFrame(rows, columns=headers)
    scraping_date = soup.find('div', class_='lastupdated').text.replace('Current as at: ', '').strip()
else:
    print("Failed to retrieve the webpage.")
```
