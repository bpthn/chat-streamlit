```python
import pandas as pd
import csv
import json
```

```python
df = pd.read_csv("ilabqa.csv")
```

```python
df.head()
```

```python
data = {}
```

```python
# Create a list to store intents
intents = []

# Iterate through each row of the dataframe
for index, row in df.iterrows():
    answer = row["Answers"]
    label = row["Labels"]
    
    # Check if the intent already exists
    intent_exists = False
    for intent in intents:
        if intent["tag"] == label:
            intent["responses"].append(answer)
            intent_exists = True
            break
    
    # If the intent doesn't exist, create a new one
    if not intent_exists:
        intents.append({"tag": label, "responses": [answer]})

# Create a dictionary with the intents list
data = {"intents": intents}

# Write the dictionary to a JSON file
with open("intents.json", "w") as f:
    json.dump(data, f, indent=4)

print("JSON file created successfully!")
```

```python
import matplotlib.pyplot as plt

# Data for overseas students
years = [2019, 2020, 2021, 2022, 2023]
student_numbers = [3800015, 3142383, 2213142, 2723047, 4156602]

# Calculate year-to-year percentage growth
growth_rates = [(student_numbers[i] - student_numbers[i - 1]) / student_numbers[i - 1] * 100 for i in range(1, len(student_numbers))]

# Create a figure with two y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the number of overseas students on the primary y-axis
ax1.plot(years, student_numbers, marker='o', linestyle='-', color='b', label='Overseas Students')
ax1.set_xlabel("Year")
ax1.set_ylabel("Number of Overseas Students", color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Highlight significant growth in 2023
ax1.plot(2023, student_numbers[-1], 'ro', label='Significant Growth')  # Red marker for 2023
ax1.annotate("Major increase in 2023", (2023, student_numbers[-1]), textcoords="offset points", xytext=(0, 10), ha='center')

# Annotate recovery trend
ax1.annotate("Recovery starts in 2022", (2022, student_numbers[-2]), textcoords="offset points", xytext=(-30, -30), ha='center', arrowprops=dict(arrowstyle="->"))

# Create a second y-axis for the percentage growth rates
ax2 = ax1.twinx()
ax2.plot(years[1:], growth_rates, 'g--', label='Year-to-Year Growth Rate')  # Dashed green line
ax2.set_ylabel("Year-to-Year Percentage Growth", color='g')
ax2.tick_params(axis='y', labelcolor='g')

# Add percentage labels on the secondary y-axis
for i, rate in enumerate(growth_rates):
    ax2.annotate(f"{rate:.2f}%", (years[i + 1], rate), textcoords="offset points", xytext=(0, 10), ha='center')

# Title and grid
plt.title("Growth of Overseas Students in Australia (2019-2023)")
plt.grid(True)

# Legends for both axes
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()  # Display the chart

```
